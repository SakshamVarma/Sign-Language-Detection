```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

data = pd.read_csv('signlangtrain.csv')

X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values
X = X / 255.0

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y = tf.keras.utils.to_categorical(y)
X = X.reshape(-1, 28, 28, 1)

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=50)

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
augmented_data = datagen.flow(X_train, y_train, batch_size=32)

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Dropout(0.3),
    Conv2D(256, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    
    Dense(y.shape[1], activation='softmax')  
])

optimizer = tf.keras.optimizers.SGD(learning_rate=0.008, momentum=0.7)
model.compile(optimizer=optimizer,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
history = model.fit(augmented_data, epochs=50, batch_size=32, validation_data=(X_val, y_val))


model.save('sign_language_model_data_aug_new3.h5')
val_loss, val_acc = model.evaluate(X_val, y_val)
print("Validation accuracy:", val_acc)

test_data = pd.read_csv('signlangtest.csv')
X_test = test_data.iloc[:, 1:].values
y_test = test_data.iloc[:, 0].values
X_test = X_test / 255.0
X_test = X_test.reshape(-1, 28, 28, 1)

y_test = label_encoder.transform(y_test)
y_test = tf.keras.utils.to_categorical(y_test)
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test accuracy:", test_acc)

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss vs. Epoch')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

test_data = pd.read_csv('signlangtest.csv')
X_test = test_data.iloc[:, 1:].values
X_test = X_test / 255.0  
X_test = X_test.reshape(-1, 28, 28, 1) 
model = load_model('sign_language_model_data_aug_new3.h5')

random_index = np.random.randint(0, len(X_test))
test_instance = X_test[random_index]
true_label = test_data.iloc[random_index, 0]

prediction = model.predict(np.expand_dims(test_instance, axis=0))
predicted_label = np.argmax(prediction)

# for skipping j
if predicted_label >= 9:
    predicted_label += 1

plt.imshow(test_instance.reshape(28, 28), cmap='gray')
plt.title(f"True Label: {true_label}, Predicted Label: {predicted_label}")
plt.axis('off')
plt.show()
```
